```python
import pandas as pd
import altair as alt
from altair import datum, expr
```

```python
df = pd.read_csv('./cases.csv')
df['date'] = pd.to_datetime(df['date'])

ws = pd.read_excel(r'./windsock.xlsx')
ws.columns = [i.lower() for i in ws.columns.to_list()]
ws['date'] = pd.to_datetime(ws['date'])
df = df.merge(ws, left_on = ['date'], right_on = ['date'], how = 'left')
df.stance.fillna('false', inplace=True)
```

```python
source = df

line = alt.Chart(source).mark_line(
    color='red',
    size=3,
    strokeWidth=3
).encode(
    x='date:T',
    y='seven_day_avg:Q'
).properties(
    width='container',
    height='container'
)

points = alt.Chart(source).mark_point().encode(
    x='date:T',
    y=alt.Y(
        'seven_day_avg:Q',
        axis=alt.Axis(title='New Cases')
    ),
    color = 'stance:N', 
    href = 'source:N',
    tooltip = ['quick take:N', 'deaths:Q', 'source:N']
).transform_filter(
    (datum.stance != 'false')
).properties(
    width='container',
    height='container'
)

line + points
```
